```python
#import packages and file
import pandas as pd
import numpy as np
tata = pd.read_excel('TATA Online Retail Data Set.xlsx')
```

```python
tata
```

```python
# checking the info.
tata.info()
```

```python
# checking for duplicates
duplicates = tata.duplicated().sum()
duplicates
```

```python
# removing duplicate values
tata = tata.drop_duplicates()
```

```python
tata
```

```python
# checking the null values for CustomerID
tata.CustomerID.isna().sum()
```

```python
# dropping the null values in CustomerID
tata = tata.dropna()
```

```python
# confirming that null values are removed
tata.CustomerID.isna().sum()
```

```python
tata.info()
```

```python
# change CustomerID datatype to object
tata_copy = tata.copy()
tata_copy['CustomerID'] = tata_copy['CustomerID'].astype(str)
     
```

```python
tata.info()
```

```python
tata
```

```python
# show the countries in the country column
tata.Country.unique()
```

```python
# remove rows that are marked unspecified
tata = tata.drop(tata[tata['Country'] == 'Unspecified'].index)
```

```python
# rechecking the country column
tata.Country.unique()
```

```python
# dropping United Kingdom as it is not needed for analysis
tata = tata[tata['Country'] != 'United Kingdom']
```

```python
# checking that United Kingdom is dropped from the dataset
tata.Country.unique()
```

```python
# checking the rows where Quantity is less than 1
tata[tata.Quantity < 1]
```

```python
# removing the rows where Quantity is less than 1
tata.drop(tata[tata.Quantity < 1].index, inplace = True)
```

```python
# rechecking the rows where Quantity is less than 1. Confirming that the rows were dropped.
tata[tata.Quantity < 1] 
```

```python
# checking the rows where UnitPrice is less than 0
tata[tata.UnitPrice < 0.001]
```

```python
# removing the rows where UnitPrice is less than 0.001
tata.drop(tata[tata.UnitPrice < 0.001].index, inplace = True)
```

```python
# rechecking the rows where UnitPrice is less than 0.001. Confirming that the rows were dropped.
tata[tata.UnitPrice < 0.001] 
```

```python
tata.info()
```

```python
# create a revenue column for analysis
tata['Revenue'] = tata.Quantity * tata.UnitPrice
tata.Revenue
```

```python
tata
```

```python
# create a profit column for analysis
tata['Profit'] = tata.Revenue - tata.UnitPrice
tata.Profit
```
